```python
#Reading path of an image from a text file
f = open("imagepath.txt", "r")
path = (f.read())
print(path)
```

```python
#Reading image using OpenCV
import cv2
import matplotlib.pyplot as plt
import numpy as np
image = cv2.imread(path)
```

```python
#Displaying image
src = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(src)
plt.show()
```

```python
#Histogram Display
hist,bins = np.histogram(image.flatten(),256,[0,256])
plt.hist(image.flatten(),256,[0,256], color = 'b')
plt.xlim([0,256])
plt.show()
```

```python
#Histogram Equalization
equ = cv2.equalizeHist(src)
hist,bins = np.histogram(equ.flatten(),256,[0,256])
plt.hist(equ.flatten(),256,[0,256], color = 'b')
plt.xlim([0,256])
plt.show()
```
